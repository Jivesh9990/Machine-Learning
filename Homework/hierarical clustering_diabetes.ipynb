```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('../datasets/diabetes.csv')
df.head()
```

```python
px.scatter_3d(df,x='BloodPressure',y='SkinThickness',z='BMI',color='Insulin')
```

```python
x=df[['BloodPressure','SkinThickness','BMI']]
```

```python
plt.style.use('seaborn') 
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(14,9))
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=4)
```

```python
df['classes']=model.fit_predict(x)
```

```python
px.scatter_3d(df,x='BloodPressure',y='Glucose',z='BMI',color='classes')
```
