```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv("../datasets/Mall_Customers.csv",index_col='CustomerID')
df.head()
```

```python
df.columns=['gender','age','income','score']
```

```python
df.head()
```

```python
x=df[['income','score']]
```

```python
plt.style.use('seaborn') # plot theme
```

# Dendogram

```python
import scipy.cluster.hierarchy as sch
```

```python
#sch.linkage(x) # u can see the cluster connection as array
```

```python
x.plot.scatter(x='income',y='score')
plt.show()
```

```python
plt.figure(figsize=(14,9))
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Distances')
plt.grid(True)
plt.show()
```

# Optimal number of cluster is 5

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=5)
```

```python
df['classes']=model.fit_predict(x)
```

```python
df
```

```python
df.plot.scatter(x='income',y='score',c='classes',cmap='cool')
plt.show()
```
