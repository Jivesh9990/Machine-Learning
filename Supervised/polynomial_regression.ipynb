```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('../datasets/Position_Salaries.csv')
```

```python
df
```

## Testing in Linear Regression Approch

```python
from sklearn.linear_model import LinearRegression
```

```python
x=df[['Level']]
y=df['Salary']
```

```python
model1=LinearRegression()
model1.fit(x,y)
model1.score(x,y)*100
```

```python
ypred=model1.predict(x)
plt.scatter(x,y,c='green')
plt.plot(x,ypred,'r-')
plt.xticks(range(1,11))
plt.show()
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=5)
Px=pf.fit_transform(x)
Px
```

```python
model1=LinearRegression()
model1.fit(Px,y)
score=model1.score(Px,y)*100
ypred=model1.predict(Px)
plt.scatter(x,y,c='green')
plt.plot(x,ypred,'r-')
plt.xticks(range(1,11))
plt.title(f'score:{score:.2f}')
plt.show()
```

```python
nX=np.array([[2]])
pX_inp=pf.transform(nX)
model1.predict(pX_inp)
```

# Example 2

```python
df2=pd.read_csv('../datasets/automobileEDA.csv')
df2[['horsepower','price']]
```

```python
x=df2[['horsepower']]
y=df2['price']
```

```python
model2=LinearRegression()
model2.fit(x,y)
score=model2.score(x,y)*100
ypred=model2.predict(x)
```

```python
plt.scatter(x,y,c='green')
plt.plot(x,ypred,'r.')
plt.title(f'score:{score:.2f}')
plt.show()
```

```python
PF=PolynomialFeatures(degree=5)
PX=PF.fit_transform(x)
```

```python
model2=LinearRegression()
model2.fit(PX,y)
score=model2.score(PX,y)*100
ypred=model2.predict(PX)
plt.scatter(x,y,c='green')
plt.plot(x,ypred,'r.')
plt.title(f'score:{score:.2f}')
plt.show()
```

```python
n_X=np.array([[2]])
pX_input=PF.transform(n_X)
model2.predict(pX_input)
```
