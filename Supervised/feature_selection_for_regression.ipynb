```python
comm_nod = df['horsepower-binned'].value_counts().idxmax()
df['horsepower-binned'].replace(np.nan,comm_nod,inplace=True)
df['horsepower-binned'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
hp_bin_enc = OneHotEncoder(drop='first')
```

```python
encoded_hpbins = hp_bin_enc.fit_transform(df[['horsepower-binned']]).toarray()
```

```python
df.drop(labels=['horsepower-binned'],axis=1,inplace=True)
```

```python
enchp = pd.DataFrame(encoded_hpbins,columns=['Low',"Medium"])
```

```python
df = pd.concat([df,enchp],axis=1)
```

```python
x=df[['engine-location','wheel-base','length','width','curb-weight','engine-size','bore','horsepower','city-mpg','highway-mpg','city-L/100km','Low','Medium']]
y=df['price']
```

```python
x
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.2)
```

```python
reg=LinearRegression()
reg.fit(xtrain,ytrain)
score=reg.score(xtest,ytest)*100
ypred=reg.predict(xtrain)
```

```python
sns.distplot(ytrain,label='original data',hist=False)
sns.distplot(ypred,label='predicted data',hist=False)
plt.title('Price prediction using Linear regression')
plt.legend()
plt.grid(True)
plt.savefig('../graphs/price_prediction.png')
```
